# <ì²«ë²ˆì§¸ ë°©ë²•>

```python
i = 0
for letter in ['A', 'B', 'C']:
    print(i, letter)
    i += 1
```

# <ë‘ë²ˆì§¸ ë°©ë²•> range()ì™€ len()

```python
letters = ['A', 'B', 'C']
for i in range(len(letters)):
    letter = letters[i]
    print(i, letter)
```

ì´ ë°©ë²•ì€ ì´ì „ ë°©ë²•ë³´ë‹¤ëŠ” ë‚˜ì•„ë³´ì´ì§€ë§Œ, íŒŒì´ì¬ ì»¤ë®¤ë‹ˆí‹°ì—ì„œëŠ” ì´ëŸ¬í•œ ì½”ë“œë¥¼ ì†Œìœ„ íŒŒì´ì¬ ë‹µì§€(Pythonic) ì•Šì•„ ë³´ì¸ë‹¤ê³  í•©ë‹ˆë‹¤.

# <ì„¸ë²ˆì§¸ ë°©ë²•> enumerate()

```python
for entry in enumerate(['A', 'B', 'C']):
    print(entry)
```

enumerate() í•¨ìˆ˜ëŠ” ê¸°ë³¸ì ìœ¼ë¡œ ì¸ë±ìŠ¤ì™€ ì›ì†Œë¡œ ì´ë£¨ì–´ì§„ í„°í”Œ(tuple)ì„ ë§Œë“¤ì–´ì¤ë‹ˆë‹¤. ë”°ë¼ì„œ ì¸ë±ìŠ¤ì™€ ì›ì†Œë¥¼ ê°ê° ë‹¤ë¥¸ ë³€ìˆ˜ì— í• ë‹¹í•˜ê³  ì‹¶ë‹¤ë©´ 'ì¸ìí’€ê¸°(unpacking)'ë¥¼ í•´ì¤˜ì•¼ í•©ë‹ˆë‹¤.

```python
for i, letter in enumerate(['A', 'B', 'C']):
    print(i, letter)
    #forë¬¸ì— ì¸ìë¥¼ i, letter ë‘ê°œë¥¼ ë„£ìœ¼ë©´ ì•Œì•„ì„œ íŠœí”Œ í˜•íƒœì—ì„œ ì–¸íŒ©í‚¹ í•´ì£¼ëŠ”ë“¯.
```

# ì‹œì‘ ì¸ë±ìŠ¤ ë°”ê¾¸ê¸°
ë£¨í”„ë¥¼ ëŒë¦¬ë‹¤ë³´ë©´ ì¸ë±ìŠ¤ë¥¼ 0ì´ ì•„ë‹ˆë¼, 1ë¡œ ì‹œì‘í•˜ê³  ì‹¶ì„ ë•Œê°€ ìˆìŠµë‹ˆë‹¤. 

ì´ëŸ´ ë•ŒëŠ” enumerate() í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•  ë•Œ start ì¸ìì— ì‹œì‘í•˜ê³  ì‹¶ì€ ìˆ«ìë¥¼ ë„˜ê¸°ë©´ ë©ë‹ˆë‹¤.

```python
for i, letter in enumerate(['A', 'B', 'C'], start=1):
    print(i, letter)
```

#enumerate ì›ë¦¬
ì§€ê¸ˆê¹Œì§€ for ë¬¸ì—ì„œ enumerate() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ëŠ” ë°©ë²•ì— ëŒ€í•´ì„œ ì•Œì•„ë´¤ëŠ”ë°ìš”. enumerate() í•¨ìˆ˜ ìì²´ê°€ ì–´ë–»ê²Œ ì‘ë™í•˜ëŠ”ì§€ ì¢€ ë” ì‚´í´ë³´ê² ìŠµë‹ˆë‹¤.

íŒŒì´ì¬ì—ì„œ for ë¬¸ì€ ë‚´ë¶€ì ìœ¼ë¡œ in ë’¤ì— ì˜¤ëŠ” ëª©ë¡ì„ ëŒ€ìƒìœ¼ë¡œ ê³„ì†í•´ì„œ next() í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ê³  ìˆë‹¤ê³  ìƒê°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ë”°ë¼ì„œ, ì¼ë°˜ ë¦¬ìŠ¤íŠ¸ë¥¼ iter() í•¨ìˆ˜ì— ë„˜ê²¨ ë°˜ë³µì(iterator)ë¡œ ë§Œë“  í›„ next() í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•´ë³´ë©´ ì›ì†Œë“¤ì´ ì°¨ë¡€ë¡œ ì–»ì–´ì§€ëŠ” ê²ƒì„ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
iter_letters = iter(['A', 'B', 'C'])

print(next(iter_letters))

print(next(iter_letters))

print(next(iter_letters))
```

```python
enumerate_letters = enumerate(['A', 'B', 'C'])
print(next(enumerate_letters))

print(next(enumerate_letters))

print(next(enumerate_letters))
```
