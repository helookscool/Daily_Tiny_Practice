```python
import requests
```

```python
req = requests.get("https://m.blog.naver.com/luvpolo/221300832833")
req.text  #크롬창에서 f12버튼 누르면 개발자도구들이 뜬다.거기의 html내용과 같은내용들이 추출된것이다. 
```

```python
from bs4 import BeautifulSoup
```

```python
req.content
req.text  #--> 이런 형태로 명령어 넣으면 못알아먹게 드니까,
```

```python
soup = BeautifulSoup(req.content, "html.parser")
```

```python
print(soup)
```

```python
span_list = soup.find("tbody").find_all("span") #tbody, span 모두 html에 있는 단어들임; 크롬 개발자창을 잘 살펴봐서 어디부터 검색해야 되는지를 잘 찾아보삼. 
soup.find("tbody").find_all("span")
```

```python
soup.find("tbody").find_all("span")[0]#-->0번째 인덱스(?) 뽑아오고
```

```python
soup.find("tbody").find_all("span")[0].text #text 명령어로 뽑으면 찾는 어구가 추출됨. 
```

```python
#위의 span에 있는 모든 한글들을 추출해서 리스트형태로 만들어보자.

for span in span_list:
    print(span.text)
```

```python
#좀더 엑기스만 추려보자
span_list2 = []
for span in span_list:
    span_list2.append(span.text)
print(span_list2)
```

```python
del span_list2[0]
span_list2
```

```python
#숫자들도 다 날리고 싶은데, 일단 숫자가 자릿수가 3자리까지 가므로 3자리 이하는 다 날려라 라고 명쳥한다.
span_list3 = []
for span in span_list2:
    if len(span) >= 4:
        span_list3.append(span)
print(span_list3)
```

```python
#이제 txt형태의 파일로 저장해 주자.
f = open('C:/kor_voca.txt','w') # 해당 폴더에 kor_voca.txt 파일을 만들어서 write 모드로 열겠다
for i in span_list3:
    f.write(i+"\n") # 단어 목록 하나 쓰고 Enter 입력, +"\n" 이렇게 덪붙이면 문장끝마다 엔터친것처럼 구분해줌
f
```

```python
f.close() # 용무를 마치고 txt파일 닫기
```
